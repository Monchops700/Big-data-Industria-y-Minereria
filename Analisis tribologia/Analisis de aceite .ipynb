```python
import os
import glob
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler, StandardScaler, RobustScaler
import joblib
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"
```

```python
file = os.path.join('C:/Users/ramon/OneDrive/Escritorio/UTFS/IMI/Computacion Aplicada/C2/', 'DataAceites.csv')
df1 = pd.read_csv(file, header=0)
attrs = {
	'var': 'df1',
	'type': type(df1).__name__,
}
[attrs]
df1
```

```python
df1 = df1.rename(columns={
	'Fierro': 'ID',
	'Cromo': 'Fierro',
	'Aluminio': 'Cromo',
	'Cobre': 'Aluminio',
	'Plomo': 'Cobre',
	'Nickel': 'Plomo',
	'Plata': 'Nickel',
	'Estano': 'Plata',
	'Titanio': 'Estaño',
	'Vanadio': 'Titanio',
	'Cadmio': 'Vanadio',
	'Manganeso': 'Cadmio',
	'Sodio': 'Manganeso',
	'Potasio': 'Sodio',
	'Silicio': 'Potasio',
	'Zinc': 'Silicio',
	'Bario': 'Zinc',
	'Boro': 'Bario',
	'Calcio': 'Boro',
	'Molibdeno': 'Calcio',
	'Magnesio': 'Molibdeno',
	'Fosforo': 'Magnesio',
	'v 40': 'Fosforo',
	'Contenido agua': 'Viscosidad 40°',
	'Índice PQ': 'Contenido agua',
	'Oxidacion': 'Indice PQ',
	'Partículas > 4um': 'Oxidacion',
	'Partículas > 6um': 'Particulas >4um',
	'Partículas > 14um': 'Particulas >6um',
	'Family': 'Particulas >14um',
	'Estado': 'Familia',
	'Diagnostic': 'Estado',
	'Comentario': 'Diagnostico',
	'Recomendacion': 'Comentario',
	'Unnamed: 34': 'Recomendacion',
})
attrs = {
	'var': 'df1',
	'type': type(df1).__name__,
}
[attrs]
df1
```

```python
df1 = df1.fillna(method='ffill')

attrs = {
	'var': 'df1',
	'type': type(df1).__name__,
}
[attrs]
df1
```

```python
df1 = df1.drop([
	'Indice PQ',
	'Fosforo',
	'Magnesio',
	'Molibdeno',
	'Calcio',
	'Boro',
	'Bario',
	'Potasio',
	'Zinc',
	'Oxidacion',
	'Manganeso',
	'Sodio',
	'Vanadio',
	'Cadmio',
	'Estaño',
	'Plata',
	'Contenido agua',
	'Nickel',
	'Comentario',
	'Titanio',
], axis='columns')
attrs = {
	'var': 'df1',
	'type': type(df1).__name__,
}
[attrs]
df1
```

```python
cond = ((df1['Estado'].ne('Normal'))
 & (df1['Familia'].ne('Shafts'))
 & (df1['Familia'].ne('Hydraulics'))
 & (df1['Familia'].ne('Gears'))
 & (df1['Diagnostico'].ne('Water Contamination'))
 & (df1['Diagnostico'].ne('Normal'))
 & (df1['Diagnostico'].ne('Oil Contamination'))
 & (df1['Diagnostico'].ne('Silica and Wear'))
 & (df1['Diagnostico'].ne('Silica ISO 4406'))
 & (df1['Estado'].ne('Anormal')))
df1 = df1[cond]
attrs = {
	'var': 'df1',
	'type': type(df1).__name__,
}
[attrs]
df1
```

```python
df1 = df1.drop([
	'Recomendacion',
	'Diagnostico',
], axis='columns')
attrs = {
	'var': 'df1',
	'type': type(df1).__name__,
}
[attrs]
df1
```

```python
df1 = df1
df1.set_index([
	'ID',
], inplace=True)
attrs = {
	'var': 'df1',
	'type': type(df1).__name__,
}
[attrs]
df1
```
